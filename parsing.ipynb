    Сделан прямолинейный парсинг с перебором страниц, здесь показано, как
    работает на примере автомобиля BMW.

```python
# импорт библиотек
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
# требуемые бренды автомобилей
pars = ['SKODA', 'AUDI', 'HONDA', 'VOLVO', 'BMW', 'NISSAN', 'INFINITI',
        'MERCEDES', 'TOYOTA', 'LEXUS', 'VOLKSWAGEN', 'MITSUBISHI']
```

```python
# функция считывающая нужные данные по ссылке
def avto(url_x):
    response1 = requests.get(url_x)
    response1.encoding = response1.apparent_encoding
    page1 = BeautifulSoup(response1.text, 'html.parser')

    dd = page1.find_all('a', class_='Link Link_color_black')
    print(dd)
    year = int(dd[0].text)  # год выпуска
    print(year)
    kuzov = dd[1].text  # кузов
    print(kuzov)
    color = dd[2].text  # цвет
    print(color)
    toplivo = dd[3].text  # вид топлива
    print(toplivo)
    price = int(page1.find(
        'span', class_='OfferPriceCaption__price').text.replace('\xa0', '')[:-1])  # цена
    probeg = int(page1.find('li', "CardInfoRow CardInfoRow_kmAge").text.replace(
        '\xa0', '').replace('Пробег', '').replace('км', ''))  # пробег
    korob = page1.find('li', class_='CardInfoRow CardInfoRow_transmission').text.replace(
        'Коробка', '')  # коробка
    dvig = float(page1.find('li', class_="CardInfoRow CardInfoRow_engine").text.split(
        ' ')[0].replace('Двигатель', ''))  # двигатель в литрах
    ls = int(page1.find(
        'li', class_="CardInfoRow CardInfoRow_engine").text.split('/')[1].replace('\xa0л.с.', ''))  # двигатель в л с
    privod = page1.find(
        'li', class_="CardInfoRow CardInfoRow_drive").text.replace('Привод', '')  # привод

    sost = page1.find('li', class_="CardInfoRow CardInfoRow_state").text.replace(
        'Состояние', '')  # состояние
    vlad = int(page1.find(
        'li', class_="CardInfoRow CardInfoRow_ownersCount").text.replace(
        'Владельцы', '')[:1])  # число владельцев
    pts = page1.find(
        'li', class_="CardInfoRow CardInfoRow_pts").text.replace('ПТС', '')
    rul = page1.find(
        'li', class_="CardInfoRow CardInfoRow_wheel").text.replace('Руль', '')  # руль
    privod = page1.find(
        'li', class_="CardInfoRow CardInfoRow_drive").text.replace('Привод', '')
    tamozh = page1.find('li', class_="CardInfoRow CardInfoRow_customs").text.replace(
        'Таможня', '')  # таможня
    model = ['skoda', url_x, year, kuzov, color, toplivo, price,
             probeg, korob, dvig, ls, privod, pts, rul, sost, tamozh]
    return model
```

```python
# цикл пролистывания страниц и ссылок на примере бренда BMW
pages = 'https://www.auto.ru/cars/bmw/all/'
data_sk = {}
c = 0
for j in range(1, 2):
    url = pages+"?page="+str(j)
    print(url)
    response = requests.get(url)
    response.encoding = response.apparent_encoding
    page = BeautifulSoup(response.text, 'html.parser')
    spisok = page.find_all('h3')
    print(spisok)
    for i in range(len(spisok)-1):
        c += 1
        name = spisok[i].text
        url1 = page.find_all(
            'a', class_="Link ListingItemTitle-module__link")[i].get('href')
        try:
            avt = avto(url1)
            avt.append(name)
            data_sk.setdefault(c, avt)
        except:
            print(c, 'ошибка запроса')
```
