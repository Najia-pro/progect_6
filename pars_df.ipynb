```python
# импорт библиотек
import pandas as pd
from collections import Counter
```

```python
# получаем датасет
df = pd.read_csv('df_car.csv')
df.shape
```

```python
df.head(5)
```

```python
# убираем дубликаты
df = df.drop_duplicates(subset=['1', '6', '16'])
df.shape
```

```python
# проверка на пустые значения
df.isnull().sum()
```

```python
# предобработаем данные, изменим название, удалим ненужные столбцы
# выделим название общее название модели: бренд и марка
```

```python
df['brand'] = df['16'].apply(lambda x: x.split(' ')[0])
```

```python
df['16'] = df['16'].apply(lambda x: x.replace('-Benz', ''))
df['16'] = df['16'].apply(lambda x: x.split(' ')[0]+' ' + x.split(' ')[1])
df['16'] = df['16'].apply(lambda x: x.split('-')[0])
```

```python
df['sell_id'] = 0
df['train'] = 1
```

```python
df = df.rename({'0': 'ненужное', '1': 'car_url', '2': 'год выпуска', '4': 'цвет', '5': 'топливо',
                '6': 'цена', '7': 'пробег', '8': 'коробка', '11': 'привод', '12': 'ПТС', '13': 'руль',
                '9': 'объем л', '10': 'лс', '15': 'таможня', '14': 'состояние', '16': 'модель',
                '3': 'кузов'}, axis=1)
```

```python
df.head()
```

```python
df['бренд'] = df['модель'].apply(lambda x: x.split(' ')[0].upper())
df = df.drop(['Unnamed: 0', 'ненужное'], axis=1)
```

```python
df = df[['кузов', 'бренд', 'цвет', 'топливо', 'пробег', 'модель',
         'год выпуска', 'sell_id', 'коробка', 'ПТС', 'привод', 'руль', 'состояние',
         'таможня', 'объем л', 'лс', 'цена', 'train']].copy()
```

```python
df['модель'] = df['модель'].apply(lambda x: x.upper())
```

```python
# запись подготовленной к обработке базы
df.to_csv('df_pars.csv')
```
